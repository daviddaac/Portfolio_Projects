```python
### We are going to keep the columns that cover until the father's info too.
#data_base['EDADPADRE'] = (data_base['EDADPADRE'].astype('int'))
data_base_col = data_base.drop(data_base.iloc[:,35:], axis = 1)
data_base_col
```

```python
data_base_col[data_base_col['EDADPADRE'].isnull()]
```

```python
data_base_col = data_base_col.fillna({'EDADPADRE':0})
data_base_col['EDADPADRE'] = data_base_col['EDADPADRE'].astype('int')
data_base_col[data_base_col['EDADPADRE'].isnull()]
```

```python
# In this analizes will only keep the data relevant abaout Mexico City. In this casem, the ENTIDADNACIMIENTO equal '9'

data_base_col = data_base_col.query('ENTIDADNACIMIENTO == 9' )
data_base_col
```

```python
# That give us a total number of 115620 newborn to analizes.

data_base_col.shape
```

### Data Cleaning
---

```python
# First at all we will delete the values in the column 'EDAD' that are 999 and 888. These are for unspecified or ignored values

data_final = data_base_col.drop(data_base_col[(data_base_col['EDAD'] > 100)].index)
data_final
```

```python
#Just to keep it clean will change the 'ENTIDADNACIMIENTO' 9 to 'Ciudad de México' 

data_final['ENTIDADNACIMIENTO'] = data_final['ENTIDADNACIMIENTO'].replace(9,'Ciudad de México')
data_final
```

```python
# Columns with values between 'No' and 'Si'

data_final.loc[:,['NACIOEXTRANJERO','SECONSIDERAINDIGENA','HABLALENGUAINDIGENA','RESIDEEXTRANJERO','VIVEHIJOANTERIOR','ATENCIONPRENATAL','SOBREVIVIOPARTO','INTERRUMPIOESTUDIOS', 'TRABAJAACTUALMENTE']]
```

```python
# Changing the values of 'Si' or 'No' based on the file SI_NO.xlsx

col_sino = ['NACIOEXTRANJERO','SECONSIDERAINDIGENA','HABLALENGUAINDIGENA','RESIDEEXTRANJERO','VIVEHIJOANTERIOR','ATENCIONPRENATAL','SOBREVIVIOPARTO','INTERRUMPIOESTUDIOS', 'TRABAJAACTUALMENTE']

data_final[col_sino] = data_final[col_sino].replace([0,1,2,8,9],['NO ESPECIFICADO','SI','NO','NO APLICA','SE IGNORA'])
data_final[col_sino] = data_final[col_sino].replace(['NO APLICA','SE IGNORA','NO ESPECIFICADO'], 'Sin Info')

for col in data_final[col_sino]:
    data_final[col] = data_final[col].apply(str.title)


data_final['TRABAJAACTUALMENTE'].unique()
```

```python
# We'll take off the value 65, 999 and 998 in MUNICIPIONACIMIENTO beacause there are not municipal names for that's codes in Mexico City.

data_final = data_final.drop(data_final[data_final['MUNICIPIONACIMIENTO'] == 65].index)
data_final = data_final.drop(data_final[(data_final['MUNICIPIONACIMIENTO'] == 999) | (data_final['MUNICIPIONACIMIENTO'] == 998)].index)
data_final
```

```python
# Getting the unique values of the column 'MUNICIPIONACIMIENTO' and with the help of the docoument CDMX_entidades.csv will replace the code number of the municipal to the actual name.

data_final.MUNICIPIONACIMIENTO.unique()
```

```python
# We are going to create a several diccionar's  with the codes values and their names to repleace it in the columns
# Also we created a dictionary to repleace the name colums of the data that has codes and names

clave = {'Unnamed: 0':'clave','Unnamed: 1':'name'}

name_mun = pd.read_excel('/Users/david.davila.a/Documents/Data Science/Ejercicios/Projects/project_birth_mexico/files/sinac_catalogos_2020_2021/MUNICIPIOS_202201 (2).xlsx')
name_mun
```

```python
name_mun = name_mun[name_mun['EFE_KEY'] == 9]
name_mun = name_mun[['CATALOG_KEY',	'MUNICIPIO']]
name_mun = name_mun.rename(columns = {'CATALOG_KEY':'clave','MUNICIPIO':'name'})
name_mun = name_mun.set_index('clave')
dict_name_mun = name_mun['name'].to_dict()
dict_name_mun
```

```python
data_final['MUNICIPIONACIMIENTO'] = data_final['MUNICIPIONACIMIENTO'].replace(dict_name_mun)
data_final['MUNICIPIONACIMIENTO'].unique()
```

```python
afi = pd.read_excel('/Users/david.davila.a/Documents/Data Science/Ejercicios/Projects/project_birth_mexico/files/sinac_catalogos_2020_2021/AFILIACION_CERTIFICADOS.xlsx')
afi
```

```python
afi = afi.drop(afi.iloc[0:3].index)
afi = afi.rename(columns = clave)
afi['clave'] = (afi['clave'].astype('int'))
afi = afi.set_index('clave')
afi_dict = afi['name'].to_dict()
afi_dict
```

```python
data_final['AFILIACION'] = data_final['AFILIACION'].replace(afi_dict)
data_final['AFILIACION'] = data_final['AFILIACION'].replace(['SE IGNORA', 'NO ESPECIFICADO','NINGUNA',0], 'No info')
data_final['AFILIACION'].unique()
```

```python
edo_conyugal = pd.read_excel('/Users/david.davila.a/Documents/Data Science/Ejercicios/Projects/project_birth_mexico/files/sinac_catalogos_2020_2021/ESTADO_CONYUGAL.xlsx')
edo_conyugal
```

```python
edo_conyugal = edo_conyugal.drop(edo_conyugal.iloc[0:3].index)
edo_conyugal = edo_conyugal.rename(columns = clave)
edo_conyugal['clave'] = edo_conyugal['clave'].astype('int')
edo_conyugal = edo_conyugal.set_index('clave')
edo_conyugal_dict = edo_conyugal['name'].to_dict()
edo_conyugal_dict
```

```python
data_final['ESTADOCONYUGAL'] = data_final['ESTADOCONYUGAL'].replace(edo_conyugal_dict)
data_final['ESTADOCONYUGAL'] = data_final['ESTADOCONYUGAL'].replace([9,'NO APLICA','SE IGNORA','NO ESPECIFICADO'], 'Sin info')
data_final['ESTADOCONYUGAL'].unique()
```

```python
entidad_resi = pd.read_excel('/Users/david.davila.a/Documents/Data Science/Ejercicios/Projects/project_birth_mexico/files/sinac_catalogos_2020_2021/ENTIDADES.xlsx')
entidad_resi
```

```python
entidad_resi = entidad_resi.drop(entidad_resi.loc[0:2].index)
entidad_resi = entidad_resi.rename(columns = clave)
entidad_resi['clave'] = (entidad_resi['clave'].astype('int'))
entidad_resi = entidad_resi.set_index('clave')
entidad_resi_dict = entidad_resi['name'].to_dict()
entidad_resi_dict
```

```python
data_final['ENTIDADRESIDENCIA'] = data_final['ENTIDADRESIDENCIA'].replace(entidad_resi_dict)
data_final['ENTIDADRESIDENCIA'].unique()
```

```python
sexo = pd.read_excel('/Users/david.davila.a/Documents/Data Science/Ejercicios/Projects/project_birth_mexico/files/sinac_catalogos_2020_2021/SEXO.xlsx')
sexo
```

```python
sexo = sexo.drop(sexo.iloc[0:3].index)
sexo = sexo.rename(columns = clave)
sexo['clave'] = sexo['clave'].astype('int')
sexo = sexo.set_index('clave')
sexo_dict = sexo['name'].to_dict()
sexo_dict
```

```python
data_final['SEXO'] = (data_final['SEXO'].replace(sexo_dict))
data_final['SEXO'] = data_final['SEXO'].replace(['No Especificado','Se Ignora'],'Sin Info')
data_final['SEXO'].unique()
```

```python
escol = pd.read_excel('/Users/david.davila.a/Documents/Data Science/Ejercicios/Projects/project_birth_mexico/files/sinac_catalogos_2020_2021/ESCOLARIDAD.xlsx')
escol
```

```python
escol = escol.drop(escol.iloc[0:3].index)
escol = escol.rename(columns = clave)
escol['clave'] = escol['clave'].astype('int')
escol = escol.set_index('clave')
escol_dict = escol['name'].to_dict()
escol_dict
```

```python
data_final['ESCOLARIDAD']= data_final['ESCOLARIDAD'].replace(escol_dict)
data_final['ESCOLARIDAD'].unique()
```

```python
# Group by 'Escolaridad'

data_final['ESCOLARIDAD'] = data_final['ESCOLARIDAD'].replace(['PRIMARIA COMPLETA','PRIMARIA INCOMPLETA','SECUNDARIA COMPLETA','SECUNDARIA INCOMPLETA','TÉCNICO TERMINAL CON SECUNDARIA COMPLETA','TÉCNICO TERMINAL CON SECUNDARIA INCOMPLETA'],'Educación Básica')
data_final['ESCOLARIDAD'] = data_final['ESCOLARIDAD'].replace(['BACHILLERATO O PREPARATORIA COMPLETA','BACHILLERATO O PREPARATORIA INCOMPLETA','TÉCNICO TERMINAL CON PREPARATORIA COMPLETA','TÉCNICO TERMINAL CON PREPARATORIA INCOMPLETA'],'Educación Media')
data_final['ESCOLARIDAD'] = data_final['ESCOLARIDAD'].replace(['LICENCIATURA O PROFESIONAL COMPLETO','LICENCIATURA O PROFESIONAL INCOMPLETO','POSGRADO INCOMPLETO','POSGRADO COMPLETO'],'Educación Superior')
data_final['ESCOLARIDAD'] = data_final['ESCOLARIDAD'].replace('NINGUNA', 'Ninguna')

# droping the data that have " 'NO ESPECIFICADO', 'SE IGNORA' "

esco_nulos = ['NO ESPECIFICADO', 'SE IGNORA']
data_final = data_final.drop(data_final.query(" ESCOLARIDAD == @esco_nulos").index)

data_final['ESCOLARIDAD'].unique()
```

```python
data_final[['ENTIDADRESIDENCIA','ENTIDADNACIMIENTO']] = data_final[['ENTIDADRESIDENCIA','ENTIDADNACIMIENTO']].replace(['MÉXICO','CIUDAD DE MÉXICO'],'Ciudad de México')
data_final
```
